<b>4.</b><br>
Write `'w'` and Read `'r'` the file we downloaded from the URL<br>We use `with`-statements so we do not have to manually close the file.<br>Otherwise, we would need to do: <br>
~~~
f.open('filename')
json.dump(res, f)
f.close()
~~~
<font color = 'blue'>Escribir `'w'` y Leer `'r'` el archivo nos descargamos del URL<br>
Usamos `with`-declaraciónes a fin de que no necesitamos cerrar manualmente el archivo.<br>De otra manera, necesitamos que hacer:</font>
~~~
f.open('filename')
json.dump(res, f)
f.close()
~~~

```python
res = r.json()
```

```python
with open('valenbici.json', 'w') as f:
    json.dump(res, f, ensure_ascii=False)
```

```python
with open('valenbici.json', 'r') as f:
    data=json.load(f)
```

<b>6.</b><br>Define empty lists for each column in our DataFrame. We will populate these lists using our data.<br><font color='blue'>Definir las listas vacías por cada columna en la DataFrame. Los llenaremos las listas con los datos.</font>

```python
X_values = []
Y_values = []
name_values = []
number_values = []
open_values = []
ticket_values = []
available_values = []
free_values = []
total_values = []
update_values = []
```

<b>7.</b><br>Iterate through each item in our data (273) and save the information in our empty lists.<br><font color = 'blue'>Iterar a cada ítem en nuestros datos (273) y guardar la informacion en las listas vacías.</font>

```python
for record in range(0, len(data['features'])):
    xx = data['features'][record]['geometry']['coordinates'][0]
    yy = data['features'][record]['geometry']['coordinates'][1]
    nombre = data['features'][record]['properties']['name']
    numero = data['features'][record]['properties']['number']
    abierto = data['features'][record]['properties']['open']
    boleta = data['features'][record]['properties']['ticket']
    disponible = data['features'][record]['properties']['available']
    feliz = data['features'][record]['properties']['free']
    cantidad = data['features'][record]['properties']['total']
    fecha = data['features'][record]['properties']['updated_at']
    
    X_values.append(xx)
    Y_values.append(yy)
    name_values.append(nombre)
    number_values.append(numero)
    open_values.append(abierto)
    ticket_values.append(boleta)
    available_values.append(disponible)
    free_values.append(feliz)
    total_values.append(cantidad)
    update_values.append(fecha)
```

<b>8.</b><br>Define a dictionary with column name as `keys` and our lists of data as `values`.<br><font color = 'blue'>Definir una diccionario con los nombres de las columnas como `keys` y las listas de datos como `values`.
~~~
diccionario{key_1 : value_1,
            key_2 : value_2,
            .
            .
            .
            key_n : value_n
            }
~~~

```python
bici_dict = {"X":X_values,
            "Y":Y_values,
            "name":name_values,
            "number":number_values,
            "open": open_values,
            "ticket":ticket_values,
            "available":available_values,
            "free":free_values,
            "total":total_values,
            "update":update_values}
```

<b>9.</b><br>
Convert the dictionary to a DataFrame.<br><font color = 'blue'>Convirtir la diccionario a una DataFrame.</font>

```python
df = pd.DataFrame(bici_dict)
```

<b>10.</b><br>Save the DataFrame to our computers.<br><font color = 'blue'>Guardar la DataFrame a nuestras computadoras.</font>

```python
df.to_json('valencia_bici.json', force_ascii=False)
```
